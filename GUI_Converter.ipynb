```python
from tkinter import *
from tkinter import messagebox

root=Tk()
root.geometry('1200x700')
root.title('Interconverter')

#Variale

number=IntVar()
bin_number=StringVar()
octal_number=StringVar()
hex_number=StringVar()

#function 

def enter():
    num=number.get()
    bi=bin(num)[2:]
    bin_number.set(bi)
    
    oc=oct(num)[2:]
    octal_number.set(oc)
    
    he=hex(num)[2:]
    hex_number.set(he)

def clear():
    number.set('')
    bin_number.set('')
    octal_number.set('')
    hex_number.set('')
    
def exit():
    ch=messagebox.askyesno('Alert','Are you sure to exit ?')
    if ch>0:
        root.destroy()

#lables,text 

number_lbl=Label(root,text='Number',font='arial 25 bold',bg='green',fg='white').grid(row=0,column=0,padx=100,pady=10)
number_txt=Entry(root,width=30,bd=5,textvariable=number).grid(row=0,column=1,padx=10,pady=10)

number_lbl=Label(root,text='Binary Number',font='arial 25 bold',bg='green',fg='white').grid(row=2,column=0,padx=100,pady=10)
number_txt=Entry(root,width=30,bd=5,textvariable=bin_number).grid(row=2,column=1,padx=10,pady=10)

number_lbl=Label(root,text='Octal Number',font='arial 25 bold',bg='green',fg='white').grid(row=4,column=0,padx=100,pady=10)
number_txt=Entry(root,width=30,bd=5,textvariable=octal_number).grid(row=4,column=1,padx=10,pady=10)

number_lbl=Label(root,text='Hexa Number',font='arial 25 bold',bg='green',fg='white').grid(row=6,column=0,padx=100,pady=10)
number_txt=Entry(root,width=30,bd=5,textvariable=hex_number).grid(row=6,column=1,padx=10,pady=10)

#buttons

enter_btn=Button(root,text=' Enter ',font='arial 20 bold',fg='red',bg='gray',command=enter).grid(row=10,column=1,padx=10,pady=10)

clear_btn=Button(root,text=' Clear ',font='arial 20 bold',fg='red',bg='gray',command=clear).grid(row=14,column=1,padx=0,pady=10)

exit_btn=Button(root,text=' Exit ',font='arial 20 bold',fg='red',bg='gray',command=exit).grid(row=14,column=2,padx=10,pady=10)

root.mainloop()
```
